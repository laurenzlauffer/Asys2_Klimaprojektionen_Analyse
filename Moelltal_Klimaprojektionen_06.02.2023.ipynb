```python
# 2

def day_of_year_intervals(init=False):
    
    if init == True:
        return 'function initialized'
    
    if init == False:
    
        pot_leap_intervals = [0]

        var = 0

        for i in range(1971,2100):
            
            if np.mod(i,4) == 0:
                var = var + 366
            else:
                var = var + 365
                    

            pot_leap_intervals.append(var)


        return pot_leap_intervals
day_of_year_intervals(init=True)
```

```python
# 4

def create_dict(voi, lower_bound = 1971, upper_bound = 2100, init=False):
    
    '''This function creates an empty dictionary for every year between 1971 and 2099. 
    Every year is the key to four lists, which correspond to the four climate variables (tavg, tmin, tmax, pr).
    Every of this climate variables is a list of three further elements. These elements are the lists,
    that correspond to the rcps (rcp26,rcp45,rcp85). Every rcp-list contains further lists, the amount depending on
    the number of climate-models that simulate for the three different rcps.
    
    dictionary = [Year][tavg,tmin,tmax,pr][rcp26,rcp45,rcp85][model_1,model_2,model_3,...model_n ]
    
    '''
    
    rcps = sort_rcms()[1]
    variables = ['tavg','tmin','tmax','pr']
    
    if init == True:
        return 'function initalized'
    
    if init == False:
        
        if voi == 0:
            var_loop_range = 3
        if voi != 0:
            var_loop_range = 1
        
        if (lower_bound == 1971) and (upper_bound == 2100):
            probe_dict = dict.fromkeys([year for year in range(lower_bound,upper_bound)])

            for year in range(min(probe_dict),max(probe_dict)+1): 
                
                    probe_dict[year] = ([[list() for y in range(3)] for x in range(var_loop_range)])

                    for var in range(var_loop_range):
                
                        for rcp in range(3): # rcp26, rcp45, rcp85

                            if rcp == 0:
                                probe_dict[year][var][rcp] = [list() for x in range(8)]   
                            if rcp == 1:
                                probe_dict[year][var][rcp] = [list() for x in range(16)]
                            if rcp == 2:
                                probe_dict[year][var][rcp] = [list() for x in range(16)]

            return probe_dict
    
        if (((lower_bound != 1971) or (upper_bound != 2100)) or ((lower_bound != 1971) and (upper_bound != 2100))):
            
            probe_dict = dict.fromkeys([year for year in range(lower_bound,upper_bound)])


            for year in range(min(probe_dict),max(probe_dict)+1):
                
                probe_dict[year] = [[list() for y in range(3)] for x in range(var_loop_range)]


                for var in range(var_loop_range):
                    for rcp in range(3): # rcp26, rcp45, rcp85
                        if rcp == 0:
                            probe_dict[year][var][rcp] = [list() for x in range(8)]
                        if rcp == 1:
                            probe_dict[year][var][rcp] = [list() for x in range(16)]
                        if rcp == 2:
                            probe_dict[year][var][rcp] = [list() for x in range(16)]
            return probe_dict

create_dict(0,init=True)
# create_dict(init=False, voi=0)
```

```python
# 5

def sort_values_by_years(voi, init=False):
    
    ''' This function returns a dictionary, where a year is represented by a key and 
    the values within a year are the respective items (as lists)
    
    '''
    
    if init == True:
        return 'function initalized'
    
    if init == False:
        
    #   FUNCTION ARGUMENTS & DEPENDENCIES:
        if voi == 0: # all variables
            probe_dict = create_dict(init=False, voi = 0)
            var_loop_range = 3
            iloc_ranges = [-4,-2,-1]
            
        if voi == 1: # tavg
            probe_dict = create_dict(init=False, voi = 1)
            var_loop_range = 1
            iloc_ranges = [-4]
            
        if voi == 2:
            probe_dict = create_dict(init=False, voi = 2)
            var_loop_range = 1
            iloc_ranges = [-2]
            
        if voi == 3: # pr
            probe_dict = create_dict(init=False, voi = 3)
            var_loop_range = 1
            iloc_ranges = [-1]  
            
        name_rcms = sort_rcms()[0]
        rcps = sort_rcms()[1]
        pot_leap_intervals = day_of_year_intervals()

    #   INITIALIZE Variables:
        index = 0 # indexing rows
        model_index = 0
        rcp_list_index = 0

    #   INITIALIZE Lists: 
        header_labels = ['D','M','Y','DoY','tavg','tmin','tmax','pr']
        rel_data_var = [] # used 'var'-for-loop
        temp_list = []
            

        for dataset in range(len(name_rcms)):

            data = pd.DataFrame(pd.read_csv('/Users/laurenz/Desktop/Asys_2/Data/Moelltal_Klimaprojektionen/'+name_rcms[dataset], 
                                        header = None, delim_whitespace=True))

            index = [x for x in range(1,len(data)+1)]
            data.columns = header_labels
            data.insert(7,'i',index)
            data = data.set_index('i')
            #data = data.loc[:,data.columns!='pr']

            day_of_year = list(data.iloc[:,3])

            if (dataset >= 0) and (dataset <= 7): # depending on different number of rcp-models
                model_index = dataset
                    
            if (dataset >= 8) and (dataset <= 23):
                if dataset == 8:
                    model_index = 0
                else:
                    model_index = dataset - 8
            if (dataset >= 24) and (dataset <= 39):
                if dataset == 24:
                    model_index = 0
                else:
                    model_index = dataset - 24

            for var in range(var_loop_range): # tavg, tmin, tmax ---> pr data excluded, due to no relevance
                
                if voi == 0:
                    var = var

                rel_data_var = list(data.iloc[:, iloc_ranges[var]]) # depending on var, which column is relevant

                for year in range(min(probe_dict), max(probe_dict)+1): # 1971 - 2100
        
                    for rcp in range(len(rcps)): # rcp26, rcp45, rcp85
                        
                        if name_rcms[dataset].find(rcps[rcp]) >= 0: # if the dataset's name matches the criterium,
#                                                                     then elements are appended to corresponding list

                            for value in range(pot_leap_intervals[year-1971],(pot_leap_intervals[year-1970])): # year-1971 = 0-130
        
                                if np.mod(year,4) == 0: # potential leap years do have 366 values
                
                                    if day_of_year[value] < 366:
                                        temp_list.append(rel_data_var[value])
                            
                                    if day_of_year[value] == 366:
                                        temp_list.append(rel_data_var[value])
                                        probe_dict[year][var][rcp][model_index] = temp_list
                                        temp_list = []
                    
                                if np.mod(year,4) != 0:
                                    if day_of_year[value] < 365:
                                        temp_list.append(rel_data_var[value])

                                    if day_of_year[value] == 365:
                                        temp_list.append(rel_data_var[value])
                                        probe_dict[year][var][rcp][model_index] = temp_list
                                        temp_list = []


        return probe_dict
    

sort_values_by_years(0,init=True)
# sort_values_by_years()
```

```python
# 7

def calculate_averages(voi, ind_val, which='both', p2_lower_bound = 2071, p2_upper_bound = 2100, rel_t_var = 1, init=False):
    
    '''This function calculates the average of days that come into question, after filtering them according
       to the indicators provided.
       For this function, only 'voi' is important, but the other arguments are necessary to be defined, as they
       are passed further to the 'filter_values'-function. For further information see 'filter_values'.
    
       voi ... variable of interest
       voi = 0 ... all variables (tavg, tmax, pr)
       voi = 1 ... only tavg
       voi = 2 ... only tmax
       voi = 3 .. only pr
       
       ind_val ... value of main indictor variable
       
       rel_t_var ... relevant temperature variable (either tavg or tmax). 
                     (It sets a condition for identifying the right variable in the var-for-loop)
                     --> which temperature var. should take into account precipitation levels
                     pre-set standard: rel_t_var = tmax
       
                     
       rel_t_var = 0 ... tavg takes also into account pr.-levels
       rel_t_var = 1 ... tmax takes also into account pr.-levels
       
       '''
    
    if init == True:
        return 'function initialized'
    
    if init == False:
        if voi == 0:
#             period_1 = filter_values(0,ind_val,'p1', rel_t_var=rel_t_var)[1]
            period_2 = filter_values(0,ind_val,'p2', p2_lower_bound=p2_lower_bound, 
                                     p2_upper_bound=p2_upper_bound, rel_t_var=rel_t_var)[1]
            average_list = create_average_list(0)
            var_loop_range = 3
        if voi != 0:
#             period_1 = filter_values(voi,ind_val,'p1', rel_t_var=rel_t_var)[1]
            period_2 = filter_values(voi,ind_val,'p2', p2_lower_bound=p2_lower_bound, 
                                     p2_upper_bound=p2_upper_bound, rel_t_var=rel_t_var)[1]
            average_list = create_average_list(voi)
            var_loop_range = 1
            
        period_1_model_temp_list = []
        period_2_model_temp_list = []
        
        year_p_2 = 0
        model_range = 0
        
            
        for var in range(var_loop_range):

            for rcp in range(3):
                if rcp == 0:
                    model_range = 8
                if (rcp == 1) or (rcp == 2):
                    model_range = 16

                for model in range(model_range):
                    period_1_model_temp_list = []
                    period_2_model_temp_list = []

                    for year_p_1 in range(1971,2000):
                        year_p_2 = year_p_1 + 100
                        year_p_2 = p2_lower_bound + (year_p_1 - 1971)

                        if (len(period_1[year_p_1][var][rcp][model]) != 0):
                            period_1_model_temp_list.append(len(period_1[year_p_1][var][rcp][model]))
            
                        if (len(period_2[year_p_2][var][rcp][model]) != 0):
                            period_2_model_temp_list.append(len(period_2[year_p_2][var][rcp][model]))
                     
                    average_list[var][rcp][model] = (np.mean(period_1_model_temp_list),np.mean(period_2_model_temp_list))
                        

    return average_list

# calculate_averages(0,26,which='both',rel_t_var=1,init=False)
# calculate_averages(1,5,p2_lower_bound=1980,p2_upper_bound=2009)
calculate_averages(0,26, init=True)
```

```python
# 8

def calculate_changes(voi, ind_val, which='both', p2_lower_bound = 2071, p2_upper_bound = 2100, rel_t_var = 1, init=False):
    
    ''' This function calculates the changes between the number of days in the lists of period_1 and
        the lists of period_2. For this function only the argument voi is important, the other arguments get passed to filter_values
        via calculate_averages.
       voi ... variable of interest
       voi = 0 ... all variables (tavg, tmax, pr)
       voi = 1 ... only tavg
       voi = 2 ... only tmax
       voi = 3 .. only pr
       
       ind_val ... value of main indictor variable
       
       rel_t_var ... relevant temperature variable (either tavg or tmax). 
                     (It sets a condition for identifying the right variable in the var-for-loop)
                     --> which temperature var. should take into account precipitation levels
                     pre-set standard: rel_t_var = tmax
       
                     
       rel_t_var = 0 ... tavg takes also into account pr.-levels
       rel_t_var = 1 ... tmax takes also into account pr.-levels
       
    '''
    if init == True:
        return 'function initialized'
    
    if init == False:
             
        if voi == 0:
            average_list = calculate_averages(0, ind_val, which=which, p2_lower_bound = p2_lower_bound, 
                                              p2_upper_bound = p2_upper_bound, rel_t_var=rel_t_var)
            tavg_changes = [list() for rcp in range(3)]
            tmax_changes = [list() for rcp in range(3)]
            pr_changes = [list() for rcp in range(3)]

            model_range = 0
            
            for var in range(3): # tavg, tmax, pr
                if var == 0:
                    var_list = tavg_changes
                if var == 1:
                    var_list = tmax_changes
                if var == 2:
                    var_list = pr_changes

                for rcp in range(3):
                    if rcp == 0:
                        model_range = 8
                    if (rcp == 1):
                        model_range = 16
                    if (rcp == 2):
                        model_range = 16

                    for model in range(model_range):
                        var_list[rcp].append(average_list[var][rcp][model][1] - average_list[var][rcp][model][0])
                        
            return tavg_changes, tmax_changes, pr_changes

        if voi != 0:
            average_list = calculate_averages(voi, ind_val, which=which, p2_lower_bound = p2_lower_bound, 
                                              p2_upper_bound = p2_upper_bound, rel_t_var=rel_t_var)
            var_list = [list() for rcp in range(3)]

            for rcp in range(3):
                if rcp == 0:
                    model_range = 8
                if (rcp == 1):
                    model_range = 16
                if (rcp == 2):
                    model_range = 16

                for model in range(model_range):
                    var_list[rcp].append(average_list[0][rcp][model][1] - average_list[0][rcp][model][0])
                    
            return var_list
        
        
            
        
        
calculate_changes(1,5,1, init=True)
```

```python
# 9

def create_boxplots(voi, ind_val, p2_lower_bound = 2071, p2_upper_bound = 2100, rel_t_var = 1, init=False, save_plot=False):
    
    ''' This function produces boxplots for the RCP 2.6, RCP 4.5 and RCP 8.5 for each variable.
    
        For this function only the argument 'voi' is important, the other arguments get passed to 'filter_values()'
        via 'calculate_averages()'' and 'calculate_changes()'.
        
        voi ... variable of interest
        voi = 0 ... all variables (tavg, tmax, pr)
        voi = 1 ... only tavg
        voi = 2 ... only tmax
        voi = 3 .. only pr
       
        ind_val ... value of main indictor variable
       
        rel_t_var ... relevant temperature variable (either tavg or tmax). 
                      (It sets a condition for identifying the right variable in the var-for-loop)
                      --> which temperature var. should take into account precipitation levels
                      pre-set standard: rel_t_var = tmax
       
                     
        rel_t_var = 0 ... tavg takes also into account pr.-levels
        rel_t_var = 1 ... tmax takes also into account pr.-levels
        
        '''
    
    if init == True:
        return 'function initialized'
    
    if init == False:
        variable_in_title = ['Tavg','Tmax','Pr',]
        indicator_in_title = []
        rcps = ['rcp26','rcp45','rcp85']
        RGB_list = ['#70A0CD','#003466','#990002']
        
        if voi == 0:
            tavg_list_changes = calculate_changes(voi, ind_val, p2_lower_bound = p2_lower_bound, 
                                                 p2_upper_bound = p2_upper_bound, 
                                                  rel_t_var = rel_t_var, init=False)[0]
            tmax_list_changes = calculate_changes(voi, ind_val, p2_lower_bound = p2_lower_bound, 
                                                 p2_upper_bound = p2_upper_bound, 
                                                  rel_t_var = rel_t_var, init=False)[1]
            pr_list_changes = calculate_changes(voi, ind_val, p2_lower_bound = p2_lower_bound, 
                                                 p2_upper_bound = p2_upper_bound, 
                                                  rel_t_var = rel_t_var, init=False)[2]
#             tavg_list_changes = liste_1[0]
#             tmax_list_changes = liste_1[1]
#             pr_list_changes = liste_1[2]
            
            var_loop_range = 3
            
            if rel_t_var == 0:
                indicator_in_title = ['T.avg > 5°C. and Pr. < 1mm', 'T.max > 26°C.', 'Pr. < 1mm']
                
            if rel_t_var == 1:
                indicator_in_title = ['T.avg ≥ 26°C.', 'T.max ≥ 5°C. and Pr. < 1mm', 'Pr. < 1mm']
                
        if voi != 0:
            var_list_changes = calculate_changes(voi, ind_val, p2_lower_bound = p2_lower_bound, 
                                                 p2_upper_bound = p2_upper_bound, rel_t_var = rel_t_var, init=False)
            var_loop_range = 1
        
        for var in range(var_loop_range):
            median_values = []
            mins = []
            maxs = []
            
            if voi != 0:
                if voi != 3:
                    indicator_in_title = variable_in_title[voi-1]+' > '+str(ind_val)+'°C.'
                if voi == 3:
                    indicator_in_title = variable_in_title[voi-1]+'<'+str(ind_val)+'mm (kg/m^2)'          

            if voi == 0:
                if var == 0:
                    var_list = tavg_list_changes
                if var == 1:
                    var_list = tmax_list_changes
                if var == 2:
                    var_list = pr_list_changes
                
            if (voi != 0) and (var == 0):
                var_list = var_list_changes

            for rcp in range(3):
                y = (var_list[rcp])
                x = np.random.normal(rcp + 1, 0.025, len(y))
                median_values.append(np.median(y))
                mins.append(int(np.amin(y)))
                maxs.append(int(np.amax(y)))
                plt.boxplot(y, positions = [rcp+1], showfliers=False, widths=0.5, showmeans=None, meanline=True)
                plt.scatter(x,y, c = RGB_list[rcp], alpha = 0.75, s=20)
                
            figure = plt.figure
            title = 'change in number of days fulfilling '+indicator_in_title[var]+ '  [1971-2000 vs.'+str(p2_lower_bound)+'-'+str(p2_upper_bound)+'].'
            plt.title('Δ in number of days fulfilling '+indicator_in_title[var]+ '  [1971-2000 vs.'+str(p2_lower_bound)+'-'+str(p2_upper_bound)+']') 
            
            plt.xticks([1,2,3],rcps)
            yticks = sorted([x for x in range(np.amin(mins),np.amax(maxs),5)])# +median_values+[0])
            plt.yticks()
            plt.yticks(median_values, minor=True, 
                       labels=['Median RCP 2.6: '+str(round(median_values[0],2))+' Tage',
                               'Median RCP 4.5: '+str(round(median_values[1],2))+' Tage',
                               'Median RCP 8.5: '+str(round(median_values[2],2))+' Tage'], color='tab:orange')
            #plt.grid(axis='y', color = 'tab:grey', alpha=0.10)
            
            plt.gca().tick_params(axis='y', which='minor', direction='in', right=True,labelright=True, labelleft=False)
            plt.xlabel('reprenstative concentration pathways (RCP)', loc='center')
            plt.ylabel('Δ in number of days')
            if save_plot == True:
                plt.savefig(title, format='png')
            plt.show()

# create_boxplots(0,5,liste_1, 1,init=False)
create_boxplots(0,5,p2_lower_bound=2071, p2_upper_bound=2100, rel_t_var=1, init=False, save_plot=False)
# create_boxplots(0,26,init=True)
```

```python
# 10

def timepath(ind_val, rel_t_var = 1, init = False):
    
    if init == True:
        return 'function initialized'
    if init == False:
        
        changes_list = []
        changes_dict = create_changes_dict()
        
        for i in range(100):
            p2_lower_bound = 1971 + i
            p2_upper_bound = 2000 + i
            
            current_changes = calculate_changes(0,ind_val,p2_lower_bound=p2_lower_bound,
                                                p2_upper_bound=p2_upper_bound, rel_t_var=rel_t_var,init=False)
            
            for var in range(3):
                for rcp in range(3):
                    for perc in range(3): # perc --> percentile
                        if perc == 0:
                            perc_val = 25
                        if perc == 1:
                            perc_val = 50
                        if perc == 2:
                            perc_val = 75

                        changes_dict[p2_lower_bound][var][rcp].append(np.percentile(current_changes[var][rcp], 
                                                                        perc_val))
        
        for var in range(3):
            for rcp in range(3):
                if rcp == 0:
                    plot_color = 'green'
                if rcp == 1:
                    plot_color = 'blue'
                if rcp == 2:
                    plot_color = 'red'
                    
                for perc in range(3):
                    if perc == 0:
                        plot_alpha = 0.2
                        y_Q1 = []
                        y_list = y_Q1
                    if perc == 1:
                        plot_alpha = 1
                        y_median = []
                        y_list = y_median
                    if perc == 2:
                        plot_alpha = 0.2
                        y_Q2 = []
                        y_list = y_Q2
                    
                    
                    for i in range(1971,2071):
                        y_list.append(changes_dict[i][var][rcp][perc])
                        
                    x = [x for x in range(len(y_list))]

                    plt.plot(x,y_list, c=plot_color, alpha=plot_alpha)
                plt.fill_between(x,y_Q1,y_Q2,color=plot_color, alpha=plot_alpha)
                    
            plt.show()

timepath(5)
```
